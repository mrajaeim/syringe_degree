```python
import matplotlib.pyplot as plt
import cv2
import random
from glob import glob
import random
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
glob("./dataset/*v0*.jpg")
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
plt.subplots_adjust(wspace=-.5)
for i,file_name in enumerate(random.choices(glob("./dataset/*v0*.jpg"), k=9)):
    # Open the image
    rand_index = file_name.split("_")[0][-4:]
    img = cv2.imread(file_name)
    # Add the image to the i-th subplot
    axes[i//3, i%3].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    # Remove the axis labels
    axes[i//3, i%3].axis('off')
    axes[i//3, i%3].set_title(f"{rand_index}")

# Display the figure
plt.show()
```

```python
images = [1,50, 100,150,200, 250]
image_order = []
image_degree= []
files_names = []
# Loop through the list of images
for i, number in enumerate(images):
    # Open the image using OpenCV
    file_name = glob(f"./dataset/{number:04d}_v0*")[0]
    deg_sub_index = (file_name.index("deg_"), file_name.index(".jpg"))
    degree = int(float(file_name[deg_sub_index[0]+4:deg_sub_index[1]]) * 100) / 100
    img = cv2.imread(file_name)
    files_names.append(file_name)
    image_order.append(number)
    image_degree.append(degree)
    
for i, path in enumerate(files_names):
    img = plt.imread(path)
    imagebox = OffsetImage(img, zoom=0.05)
    ab = AnnotationBbox(imagebox, (image_order[i], image_degree[i]), frameon=False)
    plt.gca().add_artist(ab)

# Create a scatter plot
plt.scatter(image_order, image_degree)

# Add labels and title
plt.xlabel("Image Index")
plt.ylabel("Image Degree")
plt.title("Image Degree vs. Index")

# Show the plot
plt.show()
```
